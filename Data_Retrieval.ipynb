```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
```

```python
# This function returns dataFrame based on the ticker, source and start date sent in
def getDataFrame(ticker, source, start = None):
    if start is not None:
        return wb.DataReader(ticker, data_source = source , start = start)
    
    #This else if retrieves data from alpha-vantage using timeSeries
    elif source == 'alpha-vantage' and start == None:
        from alpha_vantage.timeseries import TimeSeries
        ts = TimeSeries(key = 'R55PSMFBEK16QI1C', output_format = 'pandas')
        AAPL_av, metadata = ts.get_daily_adjusted(ticker, outputsize = 'full')
        return AAPL_av
    
    # Return Error
    else:
        raise ValueError("Invalid Input")
        return -1

# Function should be called in the format: getTheDataFrame('AAPL', 'yahoo', '2015-01-01')
```
